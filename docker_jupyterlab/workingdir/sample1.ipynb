```python
import pandas as pd
import dateparser
import re
%matplotlib inline
df=pd.read_csv('data/_forestry.csv')
temp=[]
for x in df['Quarter']:
    x=re.sub('Q1','-01-01',x)
    x=re.sub('Q2','-04-01',x)
    x=re.sub('Q3','-08-01',x)
    x=re.sub('Q4','-11-01',x)
    x=dateparser.parse(x)
    #print(x)
    temp.append(x)
df['Quarter']=temp
#df.plot('Quarter','Plantation Forests',figsize=(33,5))

import plotly.offline as py
import plotly.graph_objs as go
#data = [go.Scatter(x=df['Quarter'],y=df['Natural Forests'])]
trace_high = go.Scatter(
                x=df['Quarter'],
                y=df['Natural Forests'],
                name = "Natural Forests",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

trace_low = go.Scatter(
                x=df['Quarter'],
                y=df['Plantation Forests'],
                name = "Plantation Forests",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)
data = [trace_high,trace_low]

py.iplot(data)
```
